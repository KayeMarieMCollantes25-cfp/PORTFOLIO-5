        # Program: Vet Visit Log & Symptom Checker
##### This program can help any people in knowing the right treatments and medical history of cats. Moreover, this helps to avoid further typing errors. It will make life easy.

```python
# Simple text variables
alert = "WARNING" 
log = "  ID-67_symptom:limping_Rm305  "

# Using user input for cat's symptom
symptom = input("Enter a cat symptom: ")
symptom = str(symptom)

# Cutting off extra empty spaces from the left and right sides
log_record = log
log_record = log_record.strip()
symptom = symptom.strip()
print("Clean log record: " + log_record)

# Getting the first letter
first_letter = log_record[0]
print("First letter of log: " + first_letter)

# Cut out letters from position 2 up to position 4
part_of_log = log_record[2:4]
print("Log slice [2:4]: " + part_of_log)

# Combining the alert message and the symptom together
message = alert + "! Cat is " + symptom 
print(message)

# upper() makes text uppercase. isalpha() checks if text is only letters.
print(symptom.upper())
print(symptom.isalpha())

# Checking if two text strings match exactly
if symptom == "coughing":
    print("Action Plan: Check lungs and breathing.")
elif symptom == "limping":
    print("Action Plan: Check legs and paws.")
else: 
    print("Action Plan: Keep watching the cat.")   

# Check if the user's symptom matches what is written in the log record
if symptom in log_record:
    print("Match found! This symptom matches the database record.")

# Swapping the underscore symbol with a clean space
clean_log = log_record.replace("_", " ")
print(clean_log)

# Going through the symptom word letter by letter using a number counter
print("Looping with while:")
index = 0
while index < len(symptom): 
    print(symptom[index])
    index = index + 1
```
